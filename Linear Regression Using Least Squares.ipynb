# Linear Regression using Least Squares
 


\\[ Y = mX + c \\]  


## Finding the Error  

\\[ L(x) = \sum_{i=1}^n (y_i - p_i)^2\\]  


## Least Squares method  
  
  
\\[m = \frac{\sum_{i=1}^n (x_i - \bar x)(y_i - \bar y)}{\sum_{i=1}^n (x_i - \bar x)^2}\\]  
  
\\[c = \bar y - m\bar x\\]  

```python
# Making imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(12.0,9.0)
```

```python
# Loading data
data=pd.read_csv("E:\Machine Learning\ML from scratch\Data\data.csv")
data.head(5)
```

```python
# Preprocessing Input data

X = data.iloc[:,0].values
Y = data.iloc[:,1].values
plt.scatter(X,Y)
plt.show()
```

```python
# Building the model
X_mean = np.mean(X)
Y_mean = np.mean(Y)

num = 0
den = 0

n = len(X)

for i in range(n):
    num += (X[i] - X_mean) * (Y[i]-Y_mean)
    den += (X[i] - X_mean) ** 2
    
b1 = num / den
b0 = Y_mean -b1 * X_mean 

#printing the coefficients
print(b1,b0)
```

```python
Y_pred = b0 + b1 * X
plt.scatter(X,Y, label="Datapoints") #actual
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red', label="Linear Regression") #predicted
plt.legend()
plt.show()
```

```python
# Making prediction
# Root Mean Squared Error
rmse = 0
for i in range(n):
    Y_pred=  b0 + b1* X[i]
    rmse += (Y[i] - Y_pred) ** 2
    
rmse = np.sqrt(rmse/n)

print(rmse)
```

```python
# R^2 Score = 1- SSE/SST = 1-(Sum of Squares Error)/(Sum of Sqaures Total)

SST = 0
SSE = 0

for i in range(n):
    Y_pred = b0 + b1 * X[i]
    SST += (Y[i] - Y_mean) ** 2
    SSE += (Y[i] - Y_pred) ** 2
        
score  = 1 - (SSE / SST)

print(score)
```
